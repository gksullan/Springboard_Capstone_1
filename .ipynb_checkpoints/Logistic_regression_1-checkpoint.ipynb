```python
import pandas as pd
pd.options.display.max_columns=1000
pd.options.display.width=200
pd.options.display.min_rows=60


import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df =  pd.read_csv('data_cleaned1.csv')
df.head()
```

```python
df.info()
```

```python
df = df.select_dtypes(['number'])
```

```python
df.dropna(axis=1,inplace=True)
```

```python
var_class = df.pop('CLASS')
```

```python
df.info()
```

```python
hyperparamter_table = []
```

```python

x_train,x_test,y_train,y_test = train_test_split(df,var_class,test_size=0.10,random_state=0)
hyperparamter_table += [{'test_size': 0.10, 'random_state': 0, 'data_size': df.shape}]
```

```python
logisticRegr = LogisticRegression()
```

```python
logisticRegr.fit(x_train,y_train)
```

```python
predictions = logisticRegr.predict(x_test)
```

```python
score = logisticRegr.score(x_test,y_test)
hyperparamter_table[-1]['test_score'] = score
```

```python
print(score)
```

```python
training_score = logisticRegr.score(x_train,y_train)
hyperparamter_table[-1]['train_score'] = training_score
```

```python
print(training_score)
```

```python
df_hyp = pd.DataFrame(hyperparamter_table[-1])
df_hyp
```
