```python
#Look at numerical data where missing values can be filled in
df.SSR.value_counts(dropna=False)
df['SSR'] = df.SSR.fillna(0)
```

```python
df.EXON.value_counts(dropna=False)
```

```python
df['EXON'] = df.EXON.fillna(0)
df['INTRON'] = df.INTRON.fillna(0)
```

```python
df.STRAND.value_counts(dropna=False)
```

```python
df['STRAND'] = df.STRAND.fillna(0)
```

```python
df.SIFT.value_counts(dropna=False)
```

```python
#Fill in missing values of SIFT categorization as 'unknown'
#Use pandas.get_dummies to engineer new features for this categorical variable
df['SIFT'] = df.SIFT.fillna('unknown')
df = pd.get_dummies(df,columns=['SIFT'])
df.head()
```

```python
df['BIOTYPE'] = df.BIOTYPE.fillna('unknown')
```

```python
df.ORIGIN.value_counts(dropna=False)
```

```python
#0 is unknown origin
df['ORIGIN'] = df.ORIGIN.fillna(0)
```
